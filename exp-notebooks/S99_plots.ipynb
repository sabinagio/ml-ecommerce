<hr>

| Plot Type     | Best For | Typical Use Case | Python Libraries |
|---------------|----------|------------------|------------------|
| Table Plot    | Displaying structured data | Showing summary statistics | pandas, matplotlib |
| Line Plot     | Trends over time | Time series analysis | matplotlib, seaborn |
| Bar Plot      | Comparing categories | Sales by region | matplotlib, seaborn |
| Histogram     | Distribution of one variable | Age distribution | matplotlib, seaborn |
| Box Plot      | Detecting outliers & spread | Salary spread | seaborn |
| Violin Plot   | Distribution + density | Compare distributions between groups | seaborn |
| Scatter Plot  | Relationship between two variables | Correlation analysis | matplotlib, seaborn |
| Pair Plot     | Multiple variable relationships | Feature exploration | seaborn |
| Heatmap       | Correlation matrix | Feature correlation | seaborn |
| KDE Plot      | Smoothed distribution | Density estimation | seaborn |
| Count Plot    | Frequency of categorical values | Class imbalance check | seaborn |
